# Visualizing Earnings Based on College Majors #

In this project, we will explore how using the pandas plotting functionality along with Jupyter notebook interface allows us to explore data efficiently using visualizations. 

We will be utilizing a dataset on the job outcomes of students who graduated from college between 2010 and 2012. The original data on job outcomes was released by [American Community Survey](https://www.census.gov/programs-surveys/acs/), which conducts surveys and aggregates the data. FiveThirtyEight cleaned the dataset and released it on their [Github Repo](https://github.com/fivethirtyeight/data/tree/master/college-majors). 

Each row in the dataset represents a different major in college and contains information on gender diversity, employment rates, median salaries, and more. Below are the columns contained in the dataset:

- `Rank` - Rank by median earnings (the dataset is ordered by this column).
- `Major_code` - Major code.
- `Major` - Major description.
- `Major_category` - Category of major.
- `Total` - Total number of people with major.
- `Sample_size` - Sample size (unweighted) of full-time.
- `Men` - Male graduates.
- `Women` - Female graduates.
- `ShareWomen` - Women as share of total.
- `Employed` - Number employed.
- `Median` - Median salary of full-time, year-round workers.
- `Low_wage_jobs` - Number in low-wage service jobs.
- `Full_time` - Number employed 35 hours or more.
- `Part_time` - Number employed less than 35 hours.

Using visualizations, we will explore questions like: 

- Do students in more popular majors make more money?
- How many majors are predominantly male? Predominantly Female?
- Which category of majors have the most students?

We'll proceed to import the libraries we need and remove rows containing null values.

```python
import pandas as pd
import matplotlib as plt

%matplotlib inline

recent_grads = pd.read_csv('My Documents/DATAQUEST/Datasets/recent-grads.csv')

# initial exploration of first row and how the data is structured
print(recent_grads.iloc[0])
print(recent_grads.head())
print(recent_grads.tail())
print(recent_grads.describe())
```

```python
# looking up the number of rows in `recent_grads` 
raw_data_count = recent_grads.count()
raw_data_count
```

```python
# dropping rows containing missing values and assigning resulting dataframe back to `recent_grads`
recent_grads = recent_grads.dropna()

# looking up the updated number of rows in 'recent grads' and comparing to the original count
cleaned_data_count = recent_grads.count()
print(cleaned_data_count)
print(raw_data_count)
```

## Pandas, Scatter Plots ##

Let's generate scatter plots to explore the following relations
- `Sample_size` and `Median`
- `Sample_size` and `Unemployment_rate`
- `Full_time` and `Median`
- `ShareWomen` and `Unemployment_rate`
- `Men` and `Median`
- `Women` and `Median`

```python
recent_grads.plot(x='Sample_size', y='Median', kind='scatter', title='Sample Size vs. Median Salary')
```

```python
recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter', title='Sample Size vs. Unemployment Rate')
```

```python
recent_grads.plot(x='Full_time', y='Median', kind='scatter', title='Full Time vs. Median Salary')
```

```python
recent_grads.plot(x='ShareWomen', y ='Unemployment_rate', kind='scatter', title='Share of Women vs. Unemployment Rate')
```

```python
recent_grads.plot(x='Men', y='Median', kind='scatter', title='Men vs. Median Salary')
```

```python
recent_grads.plot(x='Women', y='Median', kind='scatter', title='Women vs. Median Salary')
```

```python
recent_grads.plot(x='ShareWomen', y='Median', kind='scatter', title='Share of Women vs. Median Salary')
```

**Do students in more popular majors make more money?**

In our first cell, we found that the mean of `median` was $40,151. When looking at the plot of `Median` against `Sample_size`, we see that majors with a sample size of 2,000 or higher tend to have a median salary near 40K. Higher salaries north of 60K tend to have a smaller sample size. This is to be expected as these majors might be more competitive and strenuous. 

**Do students that majored in subjects that were majority female make more money?**

The scatter of 'Share of Women vs. Median Salary' shows a slightly negative correlation. The majority of the plots fall in the bottom right quadrant in majors where the share of women was higher; these correspond to salaries near or below the median salary of $40,000.


**Is there any link between the number of full-time employees and median salary?**

The scatter of `Full_time` and `median` exhibits a very similar correlation to the initial plot of `median` and `sample size`. The majors with the highest volume of full-time employees have median salaries at or near the mean of $40k.

## Pandas, Scatter Matrix Plot ##

In the last 2 steps, we created individual scatter plots to visualize potential relationships between columns and histograms to visualize the distributions of individual columns. A scatter matrix plot combines both scatter plots and histograms into one grid of plots and allows us to explore potential relationships and distributions simultaneously. A scatter matrix plot consists of n by n plots on a grid, where n is the number of columns, the plots on the diagonal are histograms, and the non-diagonal plots are scatter plots.

Let's start by importing `scatter_matrix`

```python
from pandas.plotting import scatter_matrix

scatter_matrix(recent_grads[['Sample_size','Median']], figsize=(10,10))
```

```python
scatter_matrix(recent_grads[['Sample_size','Median','Unemployment_rate']], figsize=(10,10))
```

## Pandas, Bar Plots ##

Now let's compare the percentage of women (`Share_women`) from the first ten rows and last ten rows of the `recent_grads` dataframe. 

```python
recent_grads[:10].plot.bar(x='Major', y='ShareWomen', legend=False)
recent_grads[163:].plot.bar(x='Major', y='ShareWomen', legend=False)
```

```python
# looking at unemployment rate

recent_grads[:10].plot.bar(x='Major', y='Unemployment_rate', legend=False)
recent_grads[163:].plot.bar(x='Major', y='Unemployment_rate', legend=False)
```

STEM majors like Petroleum Engineering, Chemical Engineering, Actuarial Science, etc. tend to have a smaller share of women. Majors in liberal arts, however, have a higher share of women. Nearly 80% of students with majors such as Early Childhood Education and Library Science are women. 

The unemployement rates of the STEM majors ranges from 0.025% to 12%, with Nuclear Engineering being an outlier at 17.5%. 
Conversely, liberal arts majors have a higher range of unemployment - ranging from 4% to 14%.
